```python
import numpy as np 
import matplotlib.pyplot as plt
import random
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np
from PIL import Image
from easydict import EasyDict as edict
import json
from pathlib import Path
import pandas as pd
```

```python
anno_path = "/home/goldjh93/dataset/LVIS/lvis_v1_val.json"
#"/home/goldjh93/dataset/coco/annotations-3/person_keypoints_val2014.json"
#"/home/goldjh93/dataset/coco/annotations/panoptic_val2017.json"

with open(anno_path) as f:
    _anno_dict = json.load(f)
    anno_dict = edict(_anno_dict)
```

```python
_ann=pd.DataFrame(anno_dict.annotations)
_ann.head(10)
# _ann.shape
```

```python
_img=pd.DataFrame(anno_dict.images)
_img.head(10)
# _img.shape
```

```python
_cate=pd.DataFrame(anno_dict.categories)
_cate.head(10)
# _cate.shape
```

```python
join_df= pd.merge(left=_img, right=_ann, how='left', left_on=['id'],right_on=['image_id'])
join_df.head(10)
```

```python
data_tb1=join_df[['image_id','bbox','category_id']]
data_tb1.head(10)
```

```python
data_tb2= pd.merge(left=data_tb1, right=_cate, how='left', left_on=['category_id'],right_on=['id'])
data_tb2.head(10)
```

```python
final_tb=data_tb2[['image_id','bbox','name']]
final_tb.head(10)
```
